```python
from sqlalchemy import create_engine
from sqlalchemy import Column, Integer, String, Float,Date

from sqlalchemy.ext.declarative import declarative_base
Base = declarative_base()
import pandas as pd
import datetime
```

```python
#Import csv files to dataframe to begin inspecting data
dfstation = pd.read_csv('clean_hawaii_stations.csv')
dfmeas = pd.read_csv('clean_hawaii_measurements.csv')
dfstation.head()
#dfmeas.head()
```

```python
# Define Measurement and Station tables
class Measurement(Base):
    __tablename__ = 'measurement'
    id = Column(Integer, primary_key=True)
    station = Column(String)
    date = Column(String)
    prcp= Column(Float)
    tobs = Column(Integer)
    
class Station(Base):
    __tablename__ = 'station'
    id = Column(Integer, primary_key=True)
    station = Column(String)
    name = Column(String)
    latitude= Column(Float)
    longitude = Column(Float)
    elevation = Column(Float)
```

```python
Base.metadata.tables
```

```python
engine = create_engine('sqlite:///hawaii.sqlite')
```

```python
#Drop exist tables in case re-run
Base.metadata.drop_all(engine)
```

```python
Base.metadata.create_all(engine)
```

```python
from sqlalchemy.orm import Session
session = Session(engine)
```

```python
#Fetching data into tables
for i in range(len(dfmeas)):
    print(dfmeas["tobs"][i])
    session.add(Measurement(station = dfmeas["station"][i],date=dfmeas["date"][i]
                            ,prcp=dfmeas["prcp"][i],tobs= int(dfmeas["tobs"][i])))

for i in range(len(dfstation)):
    session.add(Station(station = dfstation["station"][i],name = dfstation["name"][i],latitude = dfstation["latitude"][i],
        longitude = dfstation["longitude"][i],elevation = dfstation["elevation"][i]))
```

```python
engine.execute('select * from measurement').fetchall()
engine.execute('select * from station').fetchall()
```

```python
#  use the new attribute to see the queue of data ready to go into the database
session.new
```

```python
# commit() flushes whatever remaining changes remain to the database, and commits the transaction.
session.commit()
```
