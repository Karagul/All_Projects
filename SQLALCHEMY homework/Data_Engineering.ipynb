```python
import pandas as pd
```

```python
#Import csv files to dataframe to begin inspecting data
dfs = pd.read_csv('Resources/hawaii_stations.csv')
dfm = pd.read_csv('Resources/hawaii_measurements.csv')
dfs.head()
dfm.head()
```

dfm.isnull()

```python
#Inspected Stations data found that Hawaii_stations data do not have any NAN values so we move the data inspection to Hawaii_Measurements 
dfs.isnull().any()
```

```python
#Inspected Measurement data found that Hawaii_measurements data have NAN values and only on "PRCP column" 
dfm.isnull().any()
```

```python
#Checking how many rows are Null 
nan_rows = dfm[dfm['prcp'].isnull()]
```

```python
nan_rows.count()
```

```python
#From searching PRCP stands for "Precipitation (tenths of mm)" - Rain and TOBS - "Temperature at the time of observation" - Temperature

#So begin to check on NAN values with Temperature values since the Temperature data are not Null

notnull_rows = dfm[dfm['prcp'].notnull()]
```

```python
notnull_rows 
```

```python
#Calculating average of temperature values with Null values of Rain values
nan_rows["tobs"].mean()
```

```python
#Calculating average of temperature values with not Null values of Rain values
notnull_rows["tobs"].mean()
```

```python
#Calculating average of rain values with not Null values of Rain values
print ( notnull_rows["prcp"].min(),notnull_rows["prcp"].max(),notnull_rows["prcp"].mean(),notnull_rows["tobs"].mean(),len(nan_rows)/len(notnull_rows)*100 )
```

```python
#From 3 above data inspection, I found that rain data is in range 0-11.53 plus average rain is 0.16 with 18103 values affected
# on average temp data is 72.99 and 1447 null values of rain data(rows)  affected on average temp data is 74.3 
# while the null rain data only make up 7.99% compared with total rain data so that this situation  - null values of rain  will not affect 
#much on temp data and the above describe() function 's result that with temp is 74.3 will correspond with rain value is about 0.02 
# so we can keep these rows by fill in with either mininum,0.02 or average value of rain values to have more accurate temp rather delete/drop it .

#This case I will keep and fill in with minimum value of rain datas
dfm.fillna(notnull_rows["prcp"].min(),inplace=True)
```

```python
#To consolidate again on decision keeping Null values and just assigned it with Zero values 
print(notnull_rows["prcp"].mean(),notnull_rows["tobs"].mean(),dfm["prcp"].mean(),dfm["tobs"].mean())
#So average of rain values only change less than 1% while temperature seemly not change from this decision
```

```python
#Export to new CSV files to future analyse
dfm.to_csv("clean_hawaii_measurements.csv",index=False )
dfs.to_csv("clean_hawaii_stations.csv",index=False)
```
